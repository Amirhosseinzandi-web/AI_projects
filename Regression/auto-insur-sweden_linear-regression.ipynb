```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../datasets/AutoInsurSweden.csv")
df
```

```python
train_x = df[["X"]].values
train_y = df[["Y"]].values
df["X"]
```

```python
plt.scatter(train_x, train_y , marker=".")
```

```python
model = LinearRegression()
```

```python
model.fit(train_x , train_y);
```

```python
print(f"model coef is ==> {model.coef_}")
print(f"model intercept is ===> {model.intercept_}")
```

```python
model.score(train_x , train_y)
```

```python
y_hat = model.predict(train_x)

plt.scatter(train_x, train_y , marker=".")
plt.plot(train_x , y_hat , "r-")
```
